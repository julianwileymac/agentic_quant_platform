# 04 – RL Training

Train a PPO portfolio policy and log everything to MLflow.

```python
import yaml
from pathlib import Path
from aqp.rl.trainer import train_from_config

cfg = yaml.safe_load(Path('../configs/rl/ppo_portfolio.yaml').read_text())
# Reduce timesteps for a fast notebook run
cfg['training']['total_timesteps'] = 5000
result = train_from_config(cfg, run_name='nb-ppo')
result
```
