# 08 — Out-of-Sample Predictions (AQP × ML4T Ch12)

Mirrors `08_making_out_of_sample_predictions.ipynb`. We re-fit the chosen model on the full train+valid window and generate predictions on the held-out test segment, persisting to MLflow.


```python
from pathlib import Path
import yaml
from aqp.tasks.ml_tasks import train_ml_model

cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/lightgbm_signals.yaml')).read_text())
result = train_ml_model.apply(
    kwargs=dict(
        dataset_cfg=cfg['dataset'],
        model_cfg=cfg['model'],
        run_name='ml4t_ch12_oos',
        records=cfg.get('records', []),
        register_alpha=True,
    )
)
result.get()
```
