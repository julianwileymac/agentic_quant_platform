# 11 — Intraday Model (AQP × ML4T Ch12)

Mirrors `11_intraday_model.ipynb`. Trains the `HighFreqGBDT` model on the intraday features built in notebook 10 using `configs/ml/ml4t/intraday_gbm.yaml`.


```python
from pathlib import Path
import yaml
from aqp.core.registry import build_from_config
from aqp.ml.factor_eval import information_coefficient

cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/intraday_gbm.yaml')).read_text())
ds = build_from_config(cfg['dataset'])
model = build_from_config(cfg['model']).fit(ds)
preds = model.predict(ds, segment='test')
labels = ds.prepare('test')[['label']]
print('mean IC:', information_coefficient(preds.rename('pred'), labels['label']).mean())
```
