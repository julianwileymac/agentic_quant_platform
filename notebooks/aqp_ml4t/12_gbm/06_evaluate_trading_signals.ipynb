# 06 — Evaluate Trading Signals (AQP × ML4T Ch12)

Mirrors `06_evaluate_trading_signals.ipynb`. Loads predictions from notebook 05 and computes IC time-series, IR, decile spreads, and turnover via `aqp.ml.factor_eval`.


```python
from pathlib import Path
import yaml
from aqp.core.registry import build_from_config
from aqp.ml.factor_eval import information_coefficient, ic_summary, decile_returns

cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/lightgbm_signals.yaml')).read_text())
ds = build_from_config(cfg['dataset'])
model = build_from_config(cfg['model']).fit(ds)
preds = model.predict(ds, segment='test').rename('pred')
labels = ds.prepare('test')[['label']]

ic = information_coefficient(preds, labels['label'])
summary = ic_summary(ic)
summary
```

```python
decile_returns(preds, labels['label'], n=10).tail()
```
