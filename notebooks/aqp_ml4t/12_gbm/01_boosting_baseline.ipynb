# 01 — Boosting Baseline (AQP × ML4T Ch12)

Mirrors stefan-jansen ML4T notebook `01_boosting_baseline.ipynb` using the AQP framework. We train a vanilla LightGBM regressor on Alpha158 features and look at the in-sample IC.


```python
from pathlib import Path
import yaml
from aqp.core.registry import build_from_config

cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/gbm_baseline.yaml')).read_text())
dataset = build_from_config(cfg['dataset'])
model = build_from_config(cfg['model'])
model.fit(dataset)
```

```python
from aqp.ml.factor_eval import information_coefficient
preds = model.predict(dataset, segment='test')
labels = dataset.prepare('test')[ ['label'] ]
ic = information_coefficient(preds.rename('pred'), labels['label'])
print('mean IC:', ic.mean())
```
