# 03 — Sklearn GBM Tuning Results (AQP × ML4T Ch12)

Companion to notebook 02. Reads MLflow runs and renders a comparison table of the top-10 configurations by validation IC.


```python
from aqp.config import settings
import mlflow
import pandas as pd

mlflow.set_tracking_uri(settings.mlflow_tracking_uri)
client = mlflow.tracking.MlflowClient()
exp = client.get_experiment_by_name(settings.mlflow_experiment)
if exp is None:
    raise RuntimeError('Run notebook 02 first to populate the experiment.')

runs = mlflow.search_runs(exp.experiment_id, order_by=['metrics.ic DESC'])
cols = [c for c in runs.columns if c.startswith('params.') or c.startswith('metrics.ic')]
runs[['run_id'] + cols].head(10)
```
