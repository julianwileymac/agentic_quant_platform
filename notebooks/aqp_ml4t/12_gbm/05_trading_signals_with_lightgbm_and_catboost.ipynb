# 05 — LightGBM + CatBoost Trading Signals (AQP × ML4T Ch12)

Mirrors `05_trading_signals_with_lightgbm_and_catboost.ipynb`. Trains both LightGBM and CatBoost on the same panel and writes their predictions side-by-side for downstream evaluation.


```python
from pathlib import Path
import yaml
from aqp.core.registry import build_from_config

lgb_cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/lightgbm_signals.yaml')).read_text())
cat_cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/catboost_signals.yaml')).read_text())

ds = build_from_config(lgb_cfg['dataset'])
lgb = build_from_config(lgb_cfg['model']).fit(ds)
cat = build_from_config(cat_cfg['model']).fit(ds)
```

```python
import pandas as pd
preds = pd.DataFrame({
    'lgb': lgb.predict(ds, segment='test'),
    'cat': cat.predict(ds, segment='test'),
})
preds.tail()
```
