# 10 — Intraday Features (AQP × ML4T Ch12)

Mirrors `10_intraday_features.ipynb`. Loads minute bars from the AQP DuckDB store and engineers a small intraday feature set (cumulative volume, intraday return, spread proxy).


```python
from datetime import datetime
from aqp.core.types import Symbol
from aqp.data.duckdb_engine import DuckDBHistoryProvider

syms = [Symbol(ticker=t) for t in ['SPY', 'QQQ', 'AAPL']]
bars = DuckDBHistoryProvider().get_bars(syms, datetime(2024, 1, 1), datetime(2024, 6, 30), interval='5m')
bars.head()
```

```python
import pandas as pd
df = bars.copy()
df['intraday_ret'] = df.groupby('vt_symbol')['close'].pct_change()
df['cum_vol'] = df.groupby([pd.Grouper(key='timestamp', freq='D'), 'vt_symbol'])['volume'].cumsum()
df['spread_proxy'] = (df['high'] - df['low']) / df['close']
df.tail()
```
