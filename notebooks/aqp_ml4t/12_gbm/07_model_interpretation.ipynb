# 07 — Model Interpretation (AQP × ML4T Ch12)

Mirrors `07_model_interpretation.ipynb`. We surface MDI / MDA feature importance from a fitted LightGBM model using `aqp.ml.importance`, with optional SHAP if installed.


```python
from pathlib import Path
import yaml
from aqp.core.registry import build_from_config
from aqp.ml.importance import mdi, single_feature_ic

cfg = yaml.safe_load((Path('../../../configs/ml/ml4t/lightgbm_signals.yaml')).read_text())
ds = build_from_config(cfg['dataset'])
model = build_from_config(cfg['model']).fit(ds)
imp = mdi(model)
imp.sort_values(ascending=False).head(20)
```

```python
# Optional SHAP — runs only when shap is installed.
try:
    import shap
    panel = ds.prepare('test')
    X = panel.drop(columns=[c for c in ('label', 'y') if c in panel.columns])
    explainer = shap.TreeExplainer(model.model)
    sv = explainer.shap_values(X.head(500))
    shap.summary_plot(sv, X.head(500), show=False)
except ImportError:
    print('Install shap for SHAP plots: pip install shap')
```
