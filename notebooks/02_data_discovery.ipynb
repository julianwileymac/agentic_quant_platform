# 02 – Data Discovery

Walk the Parquet lake, index metadata into ChromaDB, and query DuckDB.

```python
from aqp.data.chroma_store import ChromaStore
store = ChromaStore()
store.index_parquet_dir()
```

```python
hits = store.search_datasets('mega-cap tech stocks daily bars')
for h in hits:
    print(h['metadata'].get('vt_symbol'), '|', h['metadata'].get('path'))
```

```python
from aqp.data.duckdb_engine import get_connection
conn = get_connection()
conn.execute('SELECT vt_symbol, COUNT(*) FROM bars GROUP BY vt_symbol').fetchdf()
```

```python
from aqp.data.expressions import compute
compute('Rank($close)', conn.execute('SELECT * FROM bars LIMIT 1000').fetchdf()).head()
```
